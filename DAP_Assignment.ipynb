```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Kaggle dataset
data=pd.read_csv("https://raw.githubusercontent.com/g-shreekant/Heart-Disease-Prediction-using-Machine-Learning/master/heart.csv")
data
```

```python
'''
Data to be chosen
age , sex , oldpeak , chol
'''
```

```python
df=data.drop(['cp', 'trestbps', 'fbs', 'thalach', 'exang', 'slope', 'ca', 'thal','restecg'], axis=1)
# Note : classlabel '1' reprsents "will get a heart disease" and 0 is its inverse
df
```

```python
#1 Data visualizations taking  2 different attributes paired and the classifying hue of plot as Target
df_3d = pd.DataFrame()
df_3d['Age'] = df['age']
df_3d['Oldpeak'] = df['oldpeak']
df_3d['Chol'] = df['chol']
df_3d['Target'] = df['target']
sns.pairplot(df_3d, hue='Target')
```
